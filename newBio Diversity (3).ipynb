###Model Building

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python

# Load the dataset
# Replace 'your_dataset.csv' with the path to your dataset file
df = pd.read_csv('dataset.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Handle missing values if any
df = df.fillna(method='ffill')
```

```python
# Encode categorical variables
label_encoders = {}
categorical_columns = ['occurrenceStatus', 'habitat']
```

```python
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Encode target variable
target_encoder = LabelEncoder()
df['conservation_priority'] = target_encoder.fit_transform(df['conservation_priority'])
```

```python
# Define predictors and target variable
X = df[['occurrenceStatus', 'individualCount', 'decimalLatitude', 'decimalLongitude',
        'minimumDepthInMeters', 'maximumDepthInMeters', 'habitat']]
y = df['conservation_priority']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initialize and train the model
model = RandomForestClassifier(n_estimators=100,max_depth=2, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=target_encoder.classes_)
```

```python
print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=target_encoder.classes_, yticklabels=target_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Define function to predict conservation priority based on latitude and longitude
def predict_conservation_priority(longitude, latitude, individualCount, minimumDepthInMeters, maximumDepthInMeters, occurrenceStatus, habitat):
    # Create a DataFrame for the input with columns in the same order as X
    input_data = pd.DataFrame({
        'occurrenceStatus': [occurrenceStatus],
        'individualCount': [individualCount],
        'decimalLatitude': [latitude],
        'decimalLongitude': [longitude],
        'minimumDepthInMeters': [minimumDepthInMeters],
        'maximumDepthInMeters': [maximumDepthInMeters],
        'habitat': [habitat]
    })

    # Encode the input data using the same encoders used for training
    for col in ['occurrenceStatus', 'habitat']:
        if col in input_data.columns:
            if col in label_encoders:
                input_data[col] = label_encoders[col].transform(input_data[col])
            else:
                raise KeyError(f"Encoder for column {col} is not available.")

    # Ensure the column order matches the training data
    input_data = input_data[['occurrenceStatus', 'individualCount', 'decimalLatitude', 'decimalLongitude',
                             'minimumDepthInMeters', 'maximumDepthInMeters', 'habitat']]

    # Standardize the input data
    input_data_scaled = scaler.transform(input_data)
    
    # Make prediction
    prediction = model.predict(input_data_scaled)
    
    # Decode the prediction
    return target_encoder.inverse_transform(prediction)[0]
```

```python
# Example usage
longitude = 81.20318
latitude = 8.508009
individualCount = 2
minimumDepthInMeters = 52.0743927
maximumDepthInMeters = 189.6537603
occurrenceStatus = 'absent'  # Ensure this is a valid category
habitat = 'sandy'  # Ensure this is a valid category

predicted_priority = predict_conservation_priority(longitude, latitude, individualCount, minimumDepthInMeters, maximumDepthInMeters, occurrenceStatus, habitat)
print(f'Predicted Conservation Priority: {predicted_priority}')
```
